### Example Code Below to Calculate Cosine Simmilarity Between Two Sentences Eliminating Stop Words

```python
X ="I love horror movies"
Y ="Lights out is a horror movie"
```

```python
# X = input("Enter first string: ").lower() 
# Y = input("Enter second string: ").lower() 
X ="I love horror movies"
Y ="Lights out is a horror movie"
  
# tokenization 
X_list = word_tokenize(X)  
Y_list = word_tokenize(Y) 
  
# sw contains the list of stopwords 
sw = stopwords.words('english')  
l1 =[];l2 =[] 
  
# remove stop words from the string 
X_set = {w for w in X_list if not w in sw}  
Y_set = {w for w in Y_list if not w in sw} 
  
# form a set containing keywords of both strings  
rvector = X_set.union(Y_set)  
for w in rvector: 
    if w in X_set: l1.append(1) # create a vector 
    else: l1.append(0) 
    if w in Y_set: l2.append(1) 
    else: l2.append(0) 
c = 0
  
# cosine formula  
for i in range(len(rvector)): 
        c+= l1[i]*l2[i] 
cosine = c / float((sum(l1)*sum(l2))**0.5) 
print("similarity: ", cosine) 
```

### Cosine similarity good at identifying how simmilar one set of tokenized words are in another set.  If one set is smaller than the other and a complete subset, the cosine simmilarity is not 1.  Is it adequate to assign a cosine simmilarity score to the fit column?

Instead of using cosine similarity, use a fraction of words matched in input sentence and search sentence.  The higher the fraction, the higher the rank.

```python
fraction = []
for z in file['job_title']:
    search = "Aspiring Human Resources"
# tokenization 
    X_list = word_tokenize(z)  
    Y_list = word_tokenize(search) 
  
    # sw contains the list of stopwords 
    sw = stopwords.words('english')  
    l1 =[];l2 =[] 
  
    # remove stop words from the string 
    X_set = {w for w in X_list if not w in sw}  
    Y_set = {w for w in Y_list if not w in sw} 
  
    # form a set containing keywords of both strings  
    rvector = X_set.union(Y_set)  
    for w in rvector: 
        if w in X_set: l1.append(1) # create a vector 
        else: l1.append(0) 
        if w in Y_set: l2.append(1) 
        else: l2.append(0) 
    c = 0
  
    # Word fraction nformula  
    for i in range(len(rvector)): 
            c+= l1[i]*l2[i] 
    fraction_of_words_captured = c / len(Y_set)
    fraction.append(fraction_of_words_captured)
file['fraction_of_words'] = fraction
file.head()
```

```python
cosine_sim = []
for z in file['job_title']:
    search = "Aspiring Human Resources"
# tokenization 
    X_list = word_tokenize(z)  
    Y_list = word_tokenize(search) 
  
    # sw contains the list of stopwords 
    sw = stopwords.words('english')  
    l1 =[];l2 =[] 
  
    # remove stop words from the string 
    X_set = {w for w in X_list if not w in sw}  
    Y_set = {w for w in Y_list if not w in sw} 
  
    # form a set containing keywords of both strings  
    rvector = X_set.union(Y_set)  
    for w in rvector: 
        if w in X_set: l1.append(1) # create a vector 
        else: l1.append(0) 
        if w in Y_set: l2.append(1) 
        else: l2.append(0) 
    c = 0
  
    # Word fraction nformula  
    for i in range(len(rvector)): 
        c+= l1[i]*l2[i] 
    cosine = c / float((sum(l1)*sum(l2))**0.5) 
    cosine_sim.append(cosine)
file['cosine_sim'] = cosine_sim
file.head()
```

### Creating a LLM model to predict fitness based on Tfidf Vectorization and Linear Regression

```python
import string
def clean_text(text):
    text = text.lower()
    text = ''.join([char for char in text if char not in string.punctuation])
    return text
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

### Initiating a stemmer and lemmatizer

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
def stem_text(text):
    return ' '.join([stemmer.stem(word) for word in text.split()])
def lemmatize_text(text):
    return ' '.join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
file['cleaned_text'] = file['job_title'].apply(lambda x: clean_text(x))
```

```python
file['stemmed_text'] = file['cleaned_text'].apply(lambda x: stem_text(x))
file['lemmatized_text'] = file['stemmed_text'].apply(lambda x: lemmatize_text(x))
```

### Splitting up the data into training and test set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(file['lemmatized_text'], file['fitness'], test_size = 0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
```

### Composing LLM Model

```python
model = Pipeline([('tfidf', TfidfVectorizer()), ('clf', LinearRegression())]) # Pipeline composed of two main processes.
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

### Creating a dataframe showing y_pred and y_test

```python
dataframe = pd.DataFrame(X_test)
dataframe['y_pred'] = y_pred
dataframe['y_test'] = y_test
dataframe
```
